```python
def breath(r, c, place):
    q = queue.Queue()

    #  
    visited = [[False]*5]*5
    
    # 
    visited[r][c] = True
    
    #   
    q.put((r, c, 0)) # (r, c, distance)

    # 
    while not q.empty():
        curr = q.get()
        # curr = (r, c, distance)
        
        for i in range(4):
            nr = curr[0] + D[i][0]
            nc = curr[1] + D[i][1]
            if curr[2] > 2:
                continue
            elif nr < 0 or nc < 0 or nr > 4 or nc > 4:
                continue
            elif visited[nr][nc] == True:
                continue
            elif place[nr][nc] == 'P':
                return False
            elif place[nr][nc] == 'X':
                continue
            elif place[nr][nc] == 'O':
                q.put((nr, nc, curr[2] + 1))
                visited[nr][nc] = True
    
    return True
```

```python
from collections import deque

D = [(-1, 0), (1, 0), (0, -1), (0, 1)]
DISTANCE = 2

def search(row, col, place):
    visited = [[False] * 5] * 5
    visited[row][col] = True

    q = deque()
    q.append((row, col, 0)) #  r, c
    
    while q:
        r, c, distance = q.popleft()

        if distance == DISTANCE:
            continue
        for i in range(4):
            nr = r + D[i][0]
            nc = c + D[i][1]
            
            #  
            if 0 <= nr < 5 and 0 <= nc < 5:
                if visited[nr][nc]:
                    continue
                if place[nr][nc] == 'P':
                    return False
                elif place[nr][nc] == 'O':
                    q.append((nr, nc, distance + 1))
                    visited[nr][nc] = True
    return True
                
def check(place):
    # palce = ['PPP', 'POP', 'POX', ---]
    for r in range(5):
        for c in range(5):
            if place[r][c] == 'P':
                if search(r, c, place) == False:
                    return False
    return True


def solution(places):
    answer = []
    for place in places:
        if check(place):
            answer.append(1)
        else:
            answer.append(0)
    return answer
```

```python
from collections import deque

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]


def bfs(place, x, y):
    visited = [[False] * 5 for _ in range(5)]
    q = deque()
    q.append((x, y, 0))
    visited[x][y] = True

    while q:
        x, y, cost = q.popleft()
        if cost == 2:
            continue
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if 0 <= nx < 5 and 0 <= ny < 5:
                if visited[nx][ny]:
                    continue
                if place[nx][ny] == 'P':
                    return False
                if place[nx][ny] == 'O':
                    q.append((nx, ny, cost + 1))
                    visited[nx][ny] = True
    return True


def solution(places):
    answer = []

    for place in places:
        p = [list(place[i]) for i in range(5)]	# 
        flag = True

        for i in range(5):
            for j in range(5):
                if p[i][j] == 'P':
                    if not bfs(p, i, j):  #    
                        flag = False
            if not flag:
                break
        if flag:
            answer.append(1)
        else:
            answer.append(0)

    return answer
```
